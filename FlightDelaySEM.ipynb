```python
from random import seed
from random import randint
from random import random
import pandas as pd

seed(1)

vals = []

for iVal in range(10000):
    x1 = randint(0,1000)
    x2 = randint(0,1000)
    x3 = randint(0,1000)
    x4 = randint(0,1000)
    x5 = randint(0,1000)
    x6 = randint(0,1000)
    x7 = randint(0,1000)
    err = random()
    eta1 = 1 * x1 + 2 * x2 + 3 * x3 + 2 * x4 + 5 * x5 + 4 * x6 + err
    
    vals.append([x1, x2, x3, x4, x5, x6, x7, eta1])
    

df = pd.DataFrame(vals, columns = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'eta1'])
```
